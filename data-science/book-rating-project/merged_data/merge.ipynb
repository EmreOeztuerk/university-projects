```python
import pandas as pd
```

```python
df_goodreads = pd.read_csv('../goodreads/cleaned_goodreads_books.csv')
df_hardcover = pd.read_csv('../hardcover/cleaned_hardcover_books.csv')
```

```python
df_merged = pd.merge(df_goodreads, df_hardcover, on='map_id', how='inner')
df_merged = df_merged.drop_duplicates(subset=['map_id'], keep='first')
df_merged = df_merged[df_merged.apply(lambda row: row['author'].replace(" ", "").lower() in row['author_names'].replace(" ", "").lower(), axis = 1)]
print(len(df_merged))
print(df_merged.head(5))
```
