```python
import pandas as pd
df = pd.read_csv('all_books_ISBN.csv')
#df.head()
```

```python
genre_set = set()
for x in df['genres']:
    y = [z.strip() for z in x.replace('[','').replace(']','').replace('\'', '').split(',')]
    genre_set.update(y)
#print(genre_set)
```

```python
import os

#Fantasy Bücher
fantasy_df = df[df['genres'].str.contains('Fantasy',case=False, na=False)]
print(f"Fantasy: {len(fantasy_df)}")

if not os.path.exists('fantasy_books.csv'):
    fantasy_df.to_csv('fantasy_books.csv', index=False)

#Science Fiction Bücher
scienceFiction_df = df[df['genres'].str.contains('Science Fiction',case=False, na=False)]
print(f"Science Fiction: {len(scienceFiction_df)}")

if not os.path.exists('scienceFiction_books.csv'):
    scienceFiction_df.to_csv('scienceFiction_books.csv', index=False)

#Young Adult Bücher
youngAdult_df = df[df['genres'].str.contains('Young Adult',case=False, na=False)]
print(f"Young Adult: {len(youngAdult_df)}")

if not os.path.exists('youngAdult_books.csv'):
    youngAdult_df.to_csv('youngAdult_books.csv', index=False)

#Romance Bücher
romance_df = df[
    df['genres'].str.contains('Romance', case=False, na=False) |
    df['genres'].str.contains('Love', case=False, na=False)
]
print(f"Romance: {len(romance_df)}")

if not os.path.exists('romance_books.csv'):
    romance_df.to_csv('romance_books.csv', index=False)

#Historical Bücher
historical_df = df[
    df['genres'].str.contains('History', case=False, na=False) |
    df['genres'].str.contains('Historical', case=False, na=False)
]
print(f"Historical: {len(historical_df)}")

if not os.path.exists('historical_books.csv'):
    historical_df.to_csv('historical_books.csv', index=False)

#Horror Bücher
horror_df = df[
    df['genres'].str.contains('Horror', case=False, na=False) |
    df['genres'].str.contains('Paranormal', case=False, na=False)
]
print(f"Horror: {len(horror_df)}")

if not os.path.exists('horror_books.csv'):
    horror_df.to_csv('horror_books.csv', index=False)

#Thriller Bücher
thriller_df = df[df['genres'].str.contains('Thriller')]
print(f"Thriller: {len(thriller_df)}")

if not os.path.exists('thriller_books.csv'):
    thriller_df.to_csv('thriller.csv', index=False)

#Mytery Bücher
mystery_df = df[df['genres'].str.contains('Mystery')]
print(f"Mystery: {len(mystery_df)}")

if not os.path.exists('mystery_books.csv'):
    mystery_df.to_csv('mystery_books.csv', index=False)

#Queer Bücher
queer_df = df[
    df['genres'].str.contains('LGBT', case=False, na=False) |
    df['genres'].str.contains('Gay', case=False, na=False) |
    df['genres'].str.contains('Lesbian', case=False, na=False)|
    df['genres'].str.contains('Queer', case=False, na=False)|
    df['genres'].str.contains('Trans', case=False, na=False)
]
print(f"Queer: {len(queer_df)}")

if not os.path.exists('queer_books.csv'):
    queer_df.to_csv('queer_books.csv', index=False)

#Dystopia Bücher
dystopia_df = df[
    df['genres'].str.contains('Dystopia', case=False, na=False) |
    df['genres'].str.contains('Apocalyptic', case=False, na=False) |
    df['genres'].str.contains('Speculative', case=False, na=False)
]
print(f"Dystopia: {len(dystopia_df)}")

if not os.path.exists('dystopia_books.csv'):
    dystopia_df.to_csv('dystopia_books.csv', index=False)

#Fiction Bücher
fiction_df = df[df['genres'].str.contains('Fiction')]
print(f"Fiction: {len(fiction_df)}")

if not os.path.exists('fiction_books.csv'):
    fiction_df.to_csv('fiction_books.csv', index=False)

#Adventure Bücher
adventure_df = df[df['genres'].str.contains('Adventure')]
print(f"Adventure: {len(adventure_df)}")

if not os.path.exists('adventure_books.csv'):
    adventure_df.to_csv('adventure_books.csv', index=False)

#Classic Bücher
classic_df = df[df['genres'].str.contains('Classic')]
print(f"Classic: {len(classic_df)}")

if not os.path.exists('classic_books.csv'):
    classic_df.to_csv('classic_books.csv', index=False)

#Non-fiction Bücher
nonFiction_df = df[
    df['genres'].str.contains('Nonfiction', case=False, na=False) |
    df['genres'].str.contains('Memoir', case=False, na=False) |
    df['genres'].str.contains('Biography', case=False, na=False)
]
print(f"Non-Fiction: {len(nonFiction_df)}")

if not os.path.exists('nonFiction_books.csv'):
    nonFiction_df.to_csv('nonFiction_books.csv', index=False)


    import pandas as pd


genre_keywords = {
    "Fantasy": ["Fantasy"],
    "Science Fiction": ["Science Fiction"],
    "Young Adult": ["Young Adult"],
    "Romance": ["Romance", "Love"],
    "Historical": ["History", "Historical"],
    "Horror": ["Horror", "Paranormal"],
    "Thriller": ["Thriller"],
    "Mystery": ["Mystery"],
    "Queer": ["LGBT", "Gay", "Lesbian", "Queer", "Trans"],
    "Dystopia": ["Dystopia", "Apocalyptic", "Speculative"],
    "Fiction": ["Fiction"],
    "Adventure": ["Adventure"],
    "Classic": ["Classic"],
    "Non-Fiction": ["Nonfiction", "Memoir", "Biography"]
}

genre_groups = []

for obergenre, suchbegriffe in genre_keywords.items():
    passende_subgenres = sorted([
        g for g in genre_set
        if any(keyword.lower() in g.lower() for keyword in suchbegriffe)
    ])
    genre_groups.append({
        "Obergenre": obergenre,
        "Subgenres": ", ".join(passende_subgenres)
    })

genre_df = pd.DataFrame(genre_groups)
print(genre_df)

```
