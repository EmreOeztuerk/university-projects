### Cleaning Hardcover data to merge with Goodreads data

```python
import pandas as pd
```

Importing the hardcover data

```python
df = pd.read_csv('books_full_with_tags.csv')
print(len(df))
```

dropping all rows, where author or book-title doesn't exist, <br>
because those attributes are necessary for the mapping with the Goodreads-data

```python
df = df.dropna(subset=['author_names'])
df = df.dropna(subset=['title'])
print(len(df))
```

replacing all multiple spaces with one space, in the columns 'book_title' and 'author', <br>
and stripping the spaces in the beginning and end of those columns

```python
#replace multiple spaces with one space
df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True)
df['author_names'] = df['author_names'].str.replace(r'\s+', ' ', regex=True)

#stripping away spaces
df['title'] = df['title'].str.strip()
df['author_names'] = df['author_names'].str.strip()

df['first_author'] = df['author_names'].apply(lambda x: x.split(',',1)[0])
df['first_author'] = df['first_author'].str.strip()
```

creating and cleaning a new column for mapping


```python
#new column for mapping
df['map_id'] = df['title'] #+ df['first_author']

#deleting spaces in this col
df['map_id'] = df['map_id'].str.replace(r'\s+', '', regex=True)
```

export cleaned data as csv


```python
df.to_csv('cleaned_hardcover_books.csv', index=False)
```
