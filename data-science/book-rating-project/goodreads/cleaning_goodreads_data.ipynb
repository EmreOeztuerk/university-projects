### cleaning Goodreads-data to merge with Hardcover-data ###

```python
import pandas as pd
```

importing the goodreads csv

```python
df = pd.read_csv('goodreads_books.csv')
print(len(df))
```

dropping all rows, where author or book-title doesn't exist, <br>
because those attributes are necessary for the mapping with the Hardcover-data

```python
df = df.dropna(subset=['author'])
df = df.dropna(subset=['book_title'])
print(len(df))
```

replacing all multiple spaces with one space, in the columns 'book_title' and 'author', <br>
and stripping the spaces in the beginning and end of those columns

```python
#replace multiple spaces with one space
df['book_title'] = df['book_title'].str.replace(r'\s+', ' ', regex=True)
df['author'] = df['author'].str.replace(r'\s+', ' ', regex=True)

#stripping away spaces
df['book_title'] = df['book_title'].str.strip()
df['author'] = df['author'].str.strip()
```

creating and cleaning a new column for mapping

```python
#new column for mapping
df['map_id'] = df['book_title']# + df['author']

#deleting spaces in this col
df['map_id'] = df['map_id'].str.replace(r'\s+', '', regex=True)
```

export cleaned data as csv

```python
df.to_csv('cleaned_goodreads_books.csv', index=False)
```
